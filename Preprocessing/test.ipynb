```python
def give_path(fld_path, temp):
    dic = {}
    movies_title = []
    file_names = glob.glob(fld_path)
    for file in file_names:
        name = file.split('\\')[-1]
        genre = name.split('.')[0]
        # print('./1_movies_per_genre/' + name)
        data = pd.read_csv('./1_movies_per_genre/' + name)
        # title = data.at[:99, 'name']
        dic[genre] = data.iloc[:50, 0]  # each type of movies select 50 movie temparily
        movies_title.extend(data.iloc[:50, 0])
    movies_title = list(set(movies_title))
    return dic, movies_title
path = './1_movies_per_genre/*.csv'
movies_title_bygenres, title_list = give_path(fld_path = path, temp = True)
print(title_list)

```

```python
import re
import string
num = r'[0-9]+'
def clean_string(text):
    text = [''.join(c for c in s if c not in string.punctuation) for s in text]
    text = [word.lower() for word in text]
    text = [''.join(c for c in s if c not in num) for s in text]
    return text
cols = ['cast','director','genres','production_companies','keywords']
for c in cols:
    movies_with_meta[c] = list(map(clean_string, movies_with_meta[c]))
```
