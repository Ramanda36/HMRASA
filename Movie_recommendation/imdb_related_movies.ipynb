```python
# get related movies with target item------------------------------------------
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
wait = WebDriverWait(driver, 10)
# movies_with_related = {}
# test = {}
# m='tt0298203'
blank_id = []
wo_ori_title = []
for m in imdb_id:
  if m in ori_movies_with_related.keys():
        continue
  related = []
  #list of movies_a2014
  driver.get("https://www.imdb.com/title/" + m + "/")
  num=0
  while num < 12:
    

    related_movies = driver.find_elements_by_class_name('ipc-poster-card.ipc-poster-card--base.ipc-poster-card--dynamic-width.ipc-sub-grid-item.ipc-sub-grid-item--span-2 .ipc-lockup-overlay.ipc-focusable')
    # print(related_movies)
    # print(len(related_movies))
    wait.until(EC.visibility_of_element_located((By.CLASS_NAME,'ipc-poster-card.ipc-poster-card--base.ipc-poster-card--dynamic-width.ipc-sub-grid-item.ipc-sub-grid-item--span-2 .ipc-lockup-overlay.ipc-focusable')))
    # print(type(related_movies[0]))
    try:
      related_movies[num].click()
      try:
        # title_class = driver.find_element_by_class_name("sc-dae4a1bc-0.iTARIg")
        title_class = driver.find_element_by_class_name("sc-afe43def-3.EpHJp")
        wait.until(EC.visibility_of_element_located((By.CLASS_NAME,'sc-afe43def-3.EpHJp')))
        title = title_class.text
        related.append(title[16:])
      except:
        wo_ori_title.append(m + ',' + str((num+1)))
        try:
          # sc-b73cd867-0 gLtJub
          # sc-b73cd867-0 fpHAoP
          # title_class = driver.find_element_by_class_name("sc-b73cd867-0")
          title_class = driver.find_element_by_class_name("sc-afe43def-1.fDTGTb")
          wait.until(EC.visibility_of_element_located((By.CLASS_NAME,'sc-afe43def-1.fDTGTb')))
          title = title_class.text
          related.append(title)
        except:
          driver.get("https://www.imdb.com/title/" + m + "/")
          time.sleep(2)
          num+=1
          continue

  
      driver.get("https://www.imdb.com/title/" + m + "/")
      time.sleep(2)
      num+=1
    except IndexError:
      break
  if len(related)==0:
        blank_id.append(m)
  ori_movies_with_related[m] = related
  


```
